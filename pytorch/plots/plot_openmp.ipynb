```python
import pandas as pd
import matplotlib.pyplot as plt
import os

results_dir = "../../results"

df = pd.read_csv("../../results/linear_layer_performance.csv")

plt.plot(df["Threads"].values, df["FLOP/s"].values / 1e9, marker='o')
plt.xlabel("Number of Threads")
plt.ylabel("FLOP/s (GFLOP/s)")
plt.title("Linear layer omp performance vs Threads")
plt.grid(True)
plt.show()
print(df[df["Time(s)"] == df["Time(s)"].min()])
```
