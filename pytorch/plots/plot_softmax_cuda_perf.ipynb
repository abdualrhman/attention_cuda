```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


df_red = pd.read_csv("../../results/softmax_reduction_cuda_perf.csv")
df_crude = pd.read_csv("../../results/softmax_crude_cuda_perf.csv")

df_red = df_red[np.isfinite(df_red["GFLOPS"].values)]
df_crude = df_crude[np.isfinite(df_crude["GFLOPS"].values)]


plt.figure(figsize=(10, 6))


plt.plot(df_red["input_bytes"].values, df_red["GFLOPS"].values, marker='o', label="Softmax Reduction")
plt.plot(df_crude["input_bytes"].values, df_crude["GFLOPS"].values, marker='o', label="Softmax Crude")

theo_max = (5/8)* 2 * 1000
plt.axhline(y=theo_max, color='r', linestyle='--', label=f"Theoretical Max ({theo_max} GFLOP/s)")

# Labels and title
plt.xlabel("Input Size (bytes)")
plt.ylabel("FLOP/s (GFLOP/s)")
plt.title("CUDA Softmax Performance vs Input Size")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
